```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


data = pd.read_csv('daily_routine.csv')

#Prepare the Data
X = data.iloc[:, 1:-1].values  # Independent variables (features)
y = data.iloc[:, -1].values    # Dependent variable (target)

#Normalize the Data
def normalize_data(X):
    return (X - X.mean()) / X.std()

X_normalized = normalize_data(X)

#Initialize Parameters RANDOMLY
np.random.seed(0)
m = len(y)  # Number of training examples
n_features = X_normalized.shape[1]  # Number of features
weights = np.random.randn(n_features)
bias = np.random.randn()
```

```python
#Define Cost Function (Mean Squared Error)
def mean_squared_error(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)

#Gradient Descent
def gradient_descent(X, y, weights, bias, learning_rate, num_epochs):
    costs = []
    for epoch in range(num_epochs):
        # Forward pass
        y_pred = np.dot(X, weights) + bias

        # Compute cost
        cost = mean_squared_error(y, y_pred)
        costs.append(cost)

        # Backpropagation
        dw = (1/m) * np.dot(X.T, (y_pred - y))
        db = (1/m) * np.sum(y_pred - y)

        # Update weights and bias
        weights -= learning_rate * dw
        bias -= learning_rate * db

        # Print progress
        if (epoch+1) % 100 == 0:
            print(f'Epoch {epoch+1}/{num_epochs}, Cost: {cost}')

    return weights, bias, costs

# Define evaluation metrics
def r_squared(y_true, y_pred):
    ss_total = np.sum((y_true - np.mean(y_true)) ** 2)
    ss_residual = np.sum((y_true - y_pred) ** 2)
    r2 = 1 - (ss_residual / ss_total)
    return r2

def mean_absolute_error(y_true, y_pred):
    return np.mean(np.abs(y_true - y_pred))

def root_mean_squared_error(y_true, y_pred):
    return np.sqrt(mean_squared_error(y_true, y_pred))
```

```python
# Define hyperparameters
learning_rate = 0.01
num_epochs = np.random.randint(100, 1000)  # Randomly initialize the number of epochs

# Perform gradient descent
weights, bias, costs = gradient_descent(X_normalized, y, weights, bias, learning_rate, num_epochs)

# Calculate and print evaluation metrics..... ACCURACIES
y_pred = np.dot(X_normalized, weights) + bias
r2 = r_squared(y, y_pred)
mae = mean_absolute_error(y, y_pred)
rmse = root_mean_squared_error(y, y_pred)
print(f'R-squared: {r2}')
print(f'Mean Absolute Error: {mae}')
print(f'Root Mean Squared Error: {rmse}')
```

```python
# Plot the graph between epochs and loss
plt.plot(range(num_epochs), costs)
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Gradient Descent: Epochs vs Loss')
plt.show()
```
